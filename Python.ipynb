### 2.8.2. Sentencias de control de ciclo: break y continue.


Las sentencias de control de ciclo cambian la ejecucin de su secuencia normal. Cuando la ejecucin deja un mbito, todos los objetos automticos que se crearon en ese mbito se destruyen.

```python
vocales
```

### 2.9.2 Clases

Las clases proporcionan un medio de agrupar datos y "funcionalidad". La creacin de una nueva clase crea un nuevo tipo de objeto, lo que permite crear nuevas *instancias* de ese tipo. Cada instancia de la clase puede tener sus propios atributos. Las instancias de una clase tambin pueden tener *mtodos* (definidos por su clase) para modificar su estado.

>class Name():
>>sentencias

Los atributos son como propiedades que queremos aadir a la clase 

```python
class Auto():
  marca = ''
  color = ''
```

Creemos un objeto de la clase Auto con algunos atributos

```python
mi_auto = Auto()
mi_auto.marca = 'ferrari'
mi_auto.color = 'rojo'

print(mi_auto.marca)
print(mi_auto.color)
```

```python
mi_auto2 = Auto()
mi_auto2.marca = 'nissan'
mi_auto2.color = 'blanco'

print(mi_auto2.marca)
print(mi_auto2.color)
```

Dentro de las clases podemos definir funciones, conocidas como *mtodos* 

```python
class Persona():
  nombre = ''
  edad = ''
  
  def decir_nombre(self):
    print('Hola, mi nombre es',self.nombre)
    
  def decir_edad(self):
    print(f'Tengo {self.edad} aos')
```

```python
carolina = Persona()
carolina.nombre = 'carolina'
carolina.edad = 20
```

```python
carolina.decir_edad()
```

```python
persona2.decir_nombre()
persona2.decir_edad()
```

Los mtodos necesitan tener un argumento convenientemente llamado self, que se refiere al objeto del mtodo que est siendo llamado. Podemos pasar ms de un argumento si as lo deseamos

```python
class Persona:
    nombre = ''
    edad = ''
     
    def print_informacion(self, nombre, edad):
        print('nombre:',self.nombre)
        print('edad:',self.edad)
             
carlos = Persona()
carlos.nombre = 'carlos'
carlos.edad = '30'
carlos.print_informacion(carlos.nombre, carlos.edad)
```

La operacin de creacin de instancias crea un objeto vaco. A muchas clases les gusta crear objetos con instancias personalizadas a un estado inicial especfico. Por lo tanto, una clase puede definir un mtodo especial llamado __init __ (), como este:

```python
class Person:
  def __init__(self,n,e):
    print('Corriendo init')
    self.nombre = n
    self.edad = e
```

```python
mario = Person('mario','55')
```

```python
print('nombre:',mario.nombre)
print('edad:',mario.edad)  
```

clase para un punto cartesiano (x,y) desplazar (dx,dy)

```python
class punto():
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def desplazar(self,dx,dy):
    self.x+=dx
    self.y+=dy
  def __str__(self):
    return f'x: {self.x}, y: {self.y}'
```

```python
punto1=punto(0,0)
print(punto1)
punto1.desplazar(1,1)
print(punto1)
punto1.desplazar(1,1)
print(punto1)
```

```python
class circulo(punto):
  def __init__(self,x,y,r):
    self.x=x
    self.y=y
    self.r=r
  def __str__(self):
    return f'x: {self.x}, y: {self.y}, r: {self.r}'
```

```python
mi_circulo=circulo(0,0,1)
mi_circulo.desplazar(1,1)
mi_circulo.desplazar(1,1)
print(mi_circulo)
```
